```python
import os
import random

visa_pattern = '4' + ''.join(random.choices('0123456789', k=15))
mastercard_pattern = '5' + ''.join(random.choices('0123456789', k=15))
amex_pattern = '3' + ''.join(random.choices('0123456789', k=14))

for i in range(50):
    file_name = f'test_file_{i}.txt'
    with open(file_name, 'w') as f:
        if random.random() < 0.2:
            cc_number = random.choice([visa_pattern, mastercard_pattern, amex_pattern])
            f.write(f'This file contains a credit card number: {cc_number}\n')
        else:
            f.write('This file does not contain a credit card number.\n')

print('50 text files created!')
```

```python
import os
import zipfile
import re

with zipfile.ZipFile('text_files.zip', 'w') as zip_file:
    for i in range(50):
        file_name = f'test_file_{i}.txt'
        zip_file.write(file_name)

print('Zip file created!')

with zipfile.ZipFile('text_files.zip', 'r') as zip_file:
    zip_file.extractall('/path/to/extract/folder')

print('Files extracted!')

credit_card_pattern = re.compile(r'\b(4[0-9]{12}(?:[0-9]{3})?|5[1-5][0-9]{14}|6(?:011|5[0-9][0-9])[0-9]{12}|3[47][0-9]{13}|3(?:0[0-5]|[68][0-9])[0-9]{11}|(?:2131|1800|35\d{3})\d{11})\b')

for file_name in os.listdir('/path/to/extract/folder'):
    if file_name.endswith('.txt'):
        with open(os.path.join('/path/to/extract/folder', file_name), 'r') as f:
            contents = f.read()
            matches = credit_card_pattern.findall(contents)
            if matches:
                print(f"Credit card number found in file: {file_name}")
                print(matches)
```

```python
import os

file_objects = [(f.stat().st_ctime, f.name) for f in os.scandir('/path/to/extract/folder') if f.is_file() and f.name.endswith('.txt')]
file_objects.sort()

for _, file_name in file_objects:
    with open(os.path.join('/path/to/extract/folder', file_name), 'r') as f:
        contents = f.read()
        print(f"File: {file_name}")
        print(contents)
        print()
```
